#Challenge 1

```python
import pandas as pd


estimate_columns = [
    'Population Estimate (as of July 1) - 2010',
    'Population Estimate (as of July 1) - 2011',
    'Population Estimate (as of July 1) - 2012',
    'Population Estimate (as of July 1) - 2013',
    'Population Estimate (as of July 1) - 2014',
    'Population Estimate (as of July 1) - 2015',
    'Population Estimate (as of July 1) - 2016'
]


sum_p = pd.DataFrame(columns=['Year', 'Total'])


for column in estimate_columns:
    year = column.split('-')[-1].strip()  
    total = c1[column].sum()
    sum_p= sum_p.append({'Total': total}, ignore_index=True)


sum_p['Year'] = [int(year) for year in range(2010, 2017)]


sum_p.set_index('Year', inplace=True)


print(sum_p)
```

#Challenge 2

```python
import pandas as pd
import numpy as np

states = []
means = []
variances = []


for state in c1['Geography']:
    
    state_population = c1[c1['Geography'] == state][estimate_columns]
    
    state_mean = state_population.values.mean().astype(int)
    state_variance = state_population.values.var().astype(float)
    
    
    states.append(state)
    means.append(state_mean)
    variances.append(state_variance)


c2= pd.DataFrame({'State': states, 'Mean': means, 'Variance': variances})


largest_average_state = c2[c2['Mean'] == c2['Mean'].max()]


largest_variance_state = c2[c2['Variance'] == c2['Variance'].max()]


print("State with the largest average population:")
print(largest_average_state)

print("State with the largest variance in population:")
print(largest_variance_state)

print (c2)
```

```python

c4['Year'] = pd.to_datetime(c4['Year'], format='%m/%d/%Y')


years = c4['Year'].dt.year


unique_years = years.drop_duplicates()


formatted_years = ', '.join(str(year) for year in unique_years)


total_number_years = unique_years.count()

print(f"{formatted_years}\nTotal number of years = {total_number_years}")
```

#Challenge 6

```python
coc_500_data= c4[c4['CoC Number'].str.endswith('500')]

unique_years2 = coc_500_data['Year'].unique()
unique_states= coc_500_data['State'].unique()


years_list=[]
states_list=[]
total_homeless_count_list=[]
unsheltered_homeless_count_list = []

for year in unique_years2:
    for state in unique_states:
        c_half = coc_500_data[(coc_500_data['Year']==year) & (coc_500_data['State']==state)]
        
        total_homeless_count = c_half[c_half['Measures']== 'Total Homeless']['Count'].values[0]
        
        unsheltered_homeless_count = c_half[c_half['Measures']== 'Unsheltered Homeless']['Count'].values[0]
        
        years_list.append(year)
        states_list.append(state)
        total_homeless_count_list.append(total_homeless_count)
        unsheltered_homeless_count_list.append(unsheltered_homeless_count)
        
c6 = pd.DataFrame({
    'Year': years_list, 'State': states_list, 'Total Homeless Count': total_homeless_count_list, 'Unsheltered Homeless Count': unsheltered_homeless_count_list
})
print(c6)
```

#Challenge 7

```python

coc_500_data = c4[c4['CoC Number'].str.endswith('500')]

unique_years2 = coc_500_data['Year'].unique()

us_years_list = []
us_total_homeless_count_list = []
us_unsheltered_homeless_count_list = []

for year in unique_years2:
    c_year = coc_500_data[coc_500_data['Year'] == year]
    us_total_homeless_count = c_year[c_year['Measures']=='Total Homeless']['Count'].str.replace(',', '').astype(int).sum()
    us_unsheltered_homeless_count = c_year[c_year['Measures'] == 'Unsheltered Homeless']['Count'].str.replace(',', '').astype(int).sum()

    us_years_list.append(year)
    us_total_homeless_count_list.append(us_total_homeless_count)
    us_unsheltered_homeless_count_list.append(us_unsheltered_homeless_count)

c7 = pd.DataFrame({
    'Year': us_years_list,
    'Total Homeless Count': us_total_homeless_count_list,
    'Unsheltered Homeless Count': us_unsheltered_homeless_count_list
})
print(c7)
```

#challenge 8

```python

average_homeless_count = c7['Total Homeless Count'].mean()


std_deviation_homeless_count = c7['Total Homeless Count'].std()


max_annual_homeless_count = c7['Total Homeless Count'].max()
min_annual_homeless_count = c7['Total Homeless Count'].min()


year_max_total_homeless = c7[c7['Total Homeless Count'] == max_annual_homeless_count]['Year'].values[0]


year_min_total_homeless = c7[c7['Total Homeless Count'] == min_annual_homeless_count]['Year'].values[0]


print("Average homeless count estimated over all years:", average_homeless_count)
print("Standard deviation of total homeless counts:", std_deviation_homeless_count)
print("Maximum annual homeless count:", max_annual_homeless_count)
print("Minimum annual homeless count:", min_annual_homeless_count)
print("Year with the maximum total homeless count:", year_max_total_homeless)
print("Year with the minimum total homeless count:", year_min_total_homeless)
```

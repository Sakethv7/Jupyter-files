```python
#lab-8
import pandas as pd
from bokeh.plotting import figure, output_notebook, show, output_file
from bokeh.layouts import row, column, gridplot
from bokeh.models import ColumnDataSource, Label, LabelSet, Range1d
from math import pi
from bokeh.models.widgets import Div

_output_notebook()
#output_file("my_plot.html")

df=pd.read_csv("/Users/preethampv/Downloads/census+income 2/adult.data", header=None)
df
```

```python
def get_distribution(age1,age2,col,col_val):
    c=df[(df[0] >=age1)&(df[0] <age2) & (df[col] == col_val)].count()[0]
    return col_val,c
ages_1=[0,20,30,40,50,60,70,80,90]


def get_distvals(id):
    val_list=df[id].value_counts().index.tolist()
    return val_list
def label_fig(k):
    y=[]
    for i in range(len(k)):
        y.append(0)
    p = figure(x_range=k,height=20,width=200)
    p.circle(k,y, size=0, alpha=0)
    p.xaxis.visible = True
    p.yaxis.visible = False
    p.xaxis.major_label_orientation = pi/2
    p.toolbar.logo = None
    p.toolbar_location = None
    return p

def get_fig(id):
    col_values=get_distvals(id)
    
    fig_list=[]
    x_label=[]
    for i in range(len(ages_1)):
        x=[]
        y=[]
        for j in col_values:
            if(ages_1[i]==90):
                name,count=get_distribution(ages_1[i],100,id,j)
            else:
                name,count=get_distribution(ages_1[i],ages_1[i+1],id,j)
            x.append(name)
            y.append(count)
        ch = figure(x_range=x, width=200, height=200)
        ch.vbar(x=x, top=y, width=0.01*len(x), color='gray')
        ch.xgrid.grid_line_color=None
        ch.outline_line_color=None
        ch.toolbar.logo = None
        ch.yaxis.minor_tick_line_color = None
        ch.xaxis.visible=False
        fig_list.append(ch)
        x_label=x
    fig_list.append(label_fig(x_label))
    return column(fig_list)
    
            
chart_emp=get_fig(1)
chart_edu=get_fig(3)
chart_ms=get_fig(5)
chart_sal=get_fig(14)

divs = [Div(text=age_group,height=200) for age_group in ['lessthan 20', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-100']]
lb = column(divs)

show(row(column(lb), column(chart_ms), column(chart_sal), column(chart_edu), column(chart_emp)))

```

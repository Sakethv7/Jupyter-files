# Create a bar chart in Bokeh to visualize the Australian bushfires' affected areas over the various Australian states.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.models import NumeralTickFormatter
output_notebook()


state_names = ['Victoria', 'Queensland', 'Tasmania', 'New South Wales', 'Western Australia', 'Southern Australia']
affected_acres_data = [463000, 617000, 302000, 104000, 99000, 140000]


source = ColumnDataSource(data=dict(states=state_names, affected_acres=affected_acres_data))


plot = figure(x_range=state_names, height=400, title="Australian Bushfires Affected Areas by State", toolbar_location=None, tools="")

plot.vbar(x='states', top='affected_acres', width=0.5, source=source, legend_label="Affected Acres", line_color="white", fill_color="grey")


plot.xaxis.major_label_orientation = "vertical"


plot.xgrid.grid_line_color = None
plot.ygrid.grid_line_color = None

plot.yaxis.formatter = NumeralTickFormatter(format="0,0")


plot.y_range.start = 0
plot.y_range.end = max(affected_acres_data) + 100000  

plot.legend.title = "States"
plot.legend.label_text_font_size = "12px"


show(plot)
```

# Create a bar chart for the number of years taken by U.S. students to graduate college in 2020. The dataset looks like this:



```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource



graduation_percentages = ['35%', '19%', '20%', '10%', '9%', '7%']
graduation_durations = ['4 years', '5 years', '6 years', '7 years', '8 years', '9 years']


percentage_values = [int(percentage.strip('%')) for percentage in graduation_percentages]


source = ColumnDataSource(data=dict(percentages=graduation_percentages, durations=graduation_durations, percentage_values=percentage_values))


plot = figure(x_range=graduation_durations, height=400, title="U.S. College Graduation Durations in 2020", toolbar_location=None, tools="")


plot.vbar(x='durations', top='percentage_values', width=0.5, source=source, legend_label="Percentage", line_color="white", fill_color="grey")


plot.xaxis.major_label_orientation = "vertical"
plot.xaxis.major_label_standoff = 0

plot.xgrid.grid_line_color = None
plot.y_range.start = 0
plot.y_range.end = max(percentage_values) + 5  # Adjust the y-axis range as needed
plot.ygrid.grid_line_color = None

plot.xaxis.axis_label = "Graduation Durations"
plot.yaxis.axis_label = "Percentage of Students (%)"


show(plot)
            
```

# Create a line chart for the percentage of U.S. students that graduated college in 4 years over the time period of 2014, 205, ... 2017.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, FactorRange


graduation_years = ['2014', '2015', '2016', '2017']
graduation_percentages = [25, 27, 30, 26] 


source = ColumnDataSource(data=dict(years=graduation_years, percentages=graduation_percentages))


x_range = FactorRange(factors=graduation_years)


plot = figure(x_range=x_range, height=400, title="Percentage of U.S. Students Graduating in 4 Years (2014-2017)", toolbar_location=None, tools="")

plot.line(x='years', y='percentages', source=source, line_width=2, legend_label="4-Year Graduates", line_color="grey")

plot.xaxis.axis_label = "Year"
plot.yaxis.axis_label = "Percentage"
plot.legend.title = "Graduation"

show(plot)

```

# Create a dot plot for the performance gap between male and female high school students in different subjects, and over different years. 

```python


from bokeh.plotting import figure, show, output_notebook

output_notebook()

x_male = [82, 80, 79, 81]
y_male = ['Math-2014', 'Math-2015', 'English-2014', 'English-2015']


x_female = [83, 84, 77, 78]
y_female = ['Math-2014', 'Math-2015', 'English-2014', 'English-2015']


p = figure(y_range=y_male, width=400, height=400)

# Create circles for female and male students
p.circle(x_female, y_female, size=10, color='blue')
p.circle(x_male, y_male, size=10, color='red')


p.line(x=[84, 80], y=['Math-2015', 'Math-2015'], line_width=2, color='black')
p.line(x=[77, 79], y=['English-2014', 'English-2014'], line_width=2, color='black')
p.line(x=[78, 81], y=['English-2015', 'English-2015'], line_width=2, color='black')
p.line(x=[82, 83], y=['Math-2014', 'Math-2014'], line_width=2, color='black')


show(p)
```

```python
from bokeh.plotting import figure, show, output_notebook
from math import pi
from bokeh.models import SingleIntervalTicker, LinearAxis

output_notebook()

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts = [5, 3, 4, 2, 4, 6]

p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.y_range.start = 0

p.yaxis.minor_tick_line_color = None

p.xaxis.major_label_orientation = pi/4

ticker = SingleIntervalTicker(interval=5, num_minor_ticks=0)
xaxis = LinearAxis(ticker=ticker)
p.add_layout(xaxis, 'below')

# Corrected the typo here - use 'ticker' instead of 'Interviw'
ticker2 = SingleIntervalTicker(interval=2, num_minor_ticks=0)
yaxis = LinearAxis(ticker=ticker2)
p.add_layout(yaxis, 'left')

show(p)
```

```python
from bokeh.plotting import figure, show, output_notebook
from math import pi
from bokeh.models import LabelSet, ColumnDataSource

output_notebook()

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts = [5, 3, 4, 2, 4, 6]

p = figure(x_range=fruits, height= 450, title ='fruit Counts', 
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.y_range.start = 0

p.yaxis.minor_tick_line_color = None

p.xaxis.major_label_orientation = pi/4

src = ColumnDataSource(
    data=dict(xvals=['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries'],
              yvals=[5, 3, 4, 2, 4, 6],
              text=[5, 3, 4, 2, 4, 6]
              )
)

lbls = LabelSet(x='xvals', y='yvals', text='text', source=src, y_offset=5, x_offset=-2)

p.add_layout(lbls)
p.yaxis.visible = False

p.outline_line_color = None
show(p)
```

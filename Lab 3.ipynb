```python
from bokeh.plotting import figure, show, output_notebook

output_notebook()

fruits =['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts =[5,3,4,2,4,6]

b = figure(x_range = fruits,  height=500,  title='Fruit Counts' )
b.vbar(x=fruits, top=counts, width= 0.4, color='Orange')
# b.xgrid.grid_line_color = None
# b.y_range.start = 0

show(b)
```

```python
fp= open('names.txt', mode='r')
names=[]
ages=[]
for line in fp:

    (n,a)= line.split('\t')
    a=int(a)
    names.append(n)
    ages.append(a)

print(names, ages)
```

```python

m= figure(x_range= names, title="Names & Ages")


m.vbar(x= names, top= ages,  width=0.5, color = 'Grey')


show(m)
```
